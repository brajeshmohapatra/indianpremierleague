```python
url = 'mysql+mysqlconnector://{user}:{password}@{server}/{database}'.format(user = 'root', password = 'root', server = 'localhost',
                                                                            database = 'ipl')
engine = create_engine(url, echo = True)
base = declarative_base()
```

```python
pd.read_sql_query('show tables;', engine.connect())
```

```python
fixtures = pd.read_sql_table('fixtures', engine.connect())
fixtures.head(2)
```

```python
match_list = pd.read_sql_table('match_list', engine.connect(), columns = ['batsman', 'bowler', 'batsman_runs', 'is_wicket', 'dismissal_kind'])
```

```python
match_list
```

```python
final_matches = match_list[match_list['final'] == 1]
```

```python
most_champion = pd.DataFrame(final_matches['winner'].value_counts())
most_champion.index[0], most_champion.winner[0]
```

```python
current_champion = pd.DataFrame(final_matches[final_matches['year'] == final_matches['year'].max()])
current_champion.winner.values[0]
```

```python
current_champions_total = pd.DataFrame(final_matches[final_matches['winner'] == current_champion.winner.values[0]])
current_champions_total.shape[0]
```

```python
match_details = pd.read_sql_table('match_details', engine.connect())
```

```python
highest_runs = pd.DataFrame(match_details.groupby('batsman')['batsman_runs'].sum())
highest_runs = highest_runs.sort_values(by = 'batsman_runs', ascending = False)
highest_runs.index[0], int(highest_runs.batsman_runs[0])
```

```python
match_details[match_details['is_wicket'] == 1]['dismissal_kind'].unique()
```

```python
highest_wickets = match_details[match_details['is_wicket'] == 1]
highest_wickets = highest_wickets[~highest_wickets['dismissal_kind'].isin(['run out', 'retired hurt', 'obstructing the field'])]
```

```python
highest_wickets = pd.DataFrame(highest_wickets.groupby('bowler')['is_wicket'].sum())
highest_wickets = highest_wickets.sort_values(by = 'is_wicket', ascending = False)
highest_wickets.head()
```

```python
highest_wickets.index[0], int(highest_wickets.is_wicket[0])
```
